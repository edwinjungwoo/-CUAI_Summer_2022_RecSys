```python
import pandas as pd
import numpy as np
```

```python
characters = pd.read_csv('characters_metadata.csv')
```

```python
characters.head()
```

```python
characters['Love Count'] = characters['Love Count'].str.replace(pat = '.', repl = '', regex = True)
characters['Hate Count'] = characters['Hate Count'].str.replace(pat = '.', repl = '', regex = True)
```

```python
characters.shape
```

```python
characters.drop(characters.loc[characters['Love Count'] == 'Unknown'].index, inplace = True)
characters.drop(characters.loc[characters['Hate Count'] == 'Unknown'].index, inplace = True)
characters.drop(characters.loc[characters['Hair Color'] == 'Unknown'].index, inplace = True)

characters.head()
```

```python
characters.shape
```

```python
characters['Counts'] = characters['Love Count'].astype(float) + characters['Hate Count'].astype(float)
```

```python
characters.head()
```

```python
characters = characters.sort_values(by = 'Counts', ascending = False)
```

```python
characters['Counts'].describe()
```

```python
characters = characters.loc[characters['Counts'] > 36]
characters.tail()
```

```python
characters.shape
```

```python
characters.to_csv("characters.csv")
```
